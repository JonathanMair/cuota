```python
from cuota.importers.import_tax_data import get_income_tax_bands, get_social_security_bands, get_spanish_regimen_general
from cuota.data_classes.tax_rules import BandsGroup, TaxModel
from cuota.logic.comparators import Comparator
from cuota.logic.calculations import Calculator
```

```python
data = get_spanish_regimen_general()
```

```python
comp = Comparator(data)
```

```python
import matplotlib.pyplot as plt
```

```python
for df, year in zip(comp.dfs, comp.years):
    plt.plot(df.index, df[("Income Tax", "payable in sequence")], label=year)
plt.legend()
```

```python
cols = list(df.columns)
cols1 = {col for col, _ in cols}
cols2 = {col for _, col in cols}
cols2
```

```python
for metric in cols2:
    for rule in cols1:
        for df, year in zip(comp.dfs, comp.years):
            plt.plot(df.index, df[(rule, metric)], label=year)
            plt.title(f"{metric}: {rule}")
        plt.legend()
        plt.show()
```
