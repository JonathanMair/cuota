```python
import pickle
with open("/Users/jonathanmair/Dropbox/coding/cuota/data/social_security_tax_rules_a.pickle", "rb") as f:
    data = pickle.load(f)
```

```python
from cuota.logic.comparators import Comparator
comp = Comparator(data)
```

```python
import matplotlib.pyplot as plt
```

```python
comp.dfs
comp.years
```

```python
cols = [b for a, b in list(comp.dfs[0].columns)]
for df in comp.dfs:
    df.columns = cols
```

```python
for df, year in zip(comp.dfs, comp.years):
    x = df.index
    plt.plot(x, df["payable"], label=year)
plt.legend()
plt.title("total annual payable")
```

```python
for df, year in zip(comp.dfs, comp.years):
    x = df.index
    plt.plot(x, df["effective rate"], label=year)
plt.legend()
plt.title("total annual payable")
```
