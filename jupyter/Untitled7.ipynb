```python
from cuota.data_classes.foreign_tax_rules import UkEmployeeTaxModel
```

```python
uk = UkEmployeeTaxModel()
```

```python
uk.results(40000)
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
arr = np.linspace(10000, 100000, 500)
```

```python
sample = uk.sample(arr)
```

```python
cols = uk.df_cols()
for col_list in cols:
    for col in col_list:
        plt.plot(arr, sample[col], label=col)
    plt.legend()
    plt.show()
```
