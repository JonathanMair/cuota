```python
from cuota.data_classes.spanish_tax_rules import SpanishAutonomoModel, SpanishRegimenGeneralModel
```

```python
# get the data

aut_by_year = []
for y in range(2022, 2026, 1):
    aut_by_year.append(SpanishAutonomoModel(year=y))
```

```python
arr = range(20000, 25000, 100)
dfs_by_year = [m.sample(arr) for m in aut_by_year]
```

```python
import matplotlib.pyplot as plt
```

```python
for metric in aut_by_year[0].df_cols():
    for model, sample in zip(aut_by_year, dfs_by_year):
        plt.plot(arr, sample[metric], label=model.year)
    plt.title(metric)
    plt.legend()
    plt.show()
```
